```python
def test(a):
  URL = ('http://apis.data.go.kr/B552061/AccidentDeath/'
         'getRestTrafficAccidentDeath?'
         '&searchYear=2019'
         '&siDo=1100'
         '&guGun=1116'
         '&numOfRows=20'
         '&pageNo=1'
         '&type=json'
         '&ServiceKey=cjj2i%2FQaTX5fg4yr%2Bxtrj6a4ffRKHgCxVsoqE8x5Fh1I%2B7bsxPOCDhItnBqdcJH18QYLbMGJzh4Cs86XySCGQg%3D%3D')
  
  response = requests.get(URL)
  r = response.json()
  items = r['items']['item']

  acc_year = [item['acc_year']for item in items]
  dth_dnv_cnt = [item['dth_dnv_cnt']for item in items]
  occrrnc_dt = [item['occrrnc_dt']for item in items]
  acc_ty_cd = [item['acc_ty_cd']for item in items]
  road_frm_cd = [item['road_frm_cd']for item in items]

  acc_year = pd.Series(acc_year)
  dth_dnv_cnt = pd.Series(dth_dnv_cnt)
  occrrnc_dt = pd.Series(occrrnc_dt)
  acc_ty_cd = pd.Series(acc_ty_cd)
  road_frm_cd = pd.Series(road_frm_cd)

  df = pd.DataFrame({'acc_year':acc_year, 'dth_dnv_cnt':dth_dnv_cnt, 'occrrnc_dt':occrrnc_dt, 'acc_ty_cd':acc_ty_cd, 'road_frm_cd':road_frm_cd})
  return df
```
